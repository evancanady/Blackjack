```python
import pandas as pd
import pickle
import blackjack
```

```python
# run the Blackjack simulation
deck_count = 2
num_players = 3
num_rounds = 1000
cut_cards = True
dealer_hit_to = 17
hit_strategies = {0: {2: 13, 3: 13, 4: 12, 5: 12, 6: 12, 7: 12, 8: 17, 9: 17, 10: 17, 11: 17},
				  1: {2: 13, 3: 13, 4: 12, 5: 12, 6: 12, 7: 12, 8: 17, 9: 17, 10: 17, 11: 17},
				  2: {2: 13, 3: 13, 4: 12, 5: 12, 6: 12, 7: 12, 8: 17, 9: 17, 10: 17, 11: 17}}

blackjack.game(deck_count, num_players, num_rounds, cut_cards, dealer_hit_to, hit_strategies)
```

```python
# import the pickled dataframe
infile = open('blackjack.pickle','rb')
df = pickle.load(infile)
infile.close()
```
